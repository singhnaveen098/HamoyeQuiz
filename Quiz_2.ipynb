```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

# Reading Data

```python
df = pd.read_csv('energydata_complete.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

# Cleaning data

```python
# checking for null values
df[df.columns].isnull().sum()
```

```python
# since data is not null we have to check for duplicate data
df.duplicated().any()
```

```python
# Dropping useless columns from data
df = df.drop(['date', 'lights'], axis=1)
df
```

# We Can Normalize the data for better results using min max scaler

```python
scaler = MinMaxScaler()
df[df.columns] = scaler.fit_transform(df)
df
```

# Question 12

```python
x2 = np.array(df['T2'])
y2 = np.array(df['T6'])
#we have to reshape x2 because of only one predictor
x2 = x2.reshape(-1, 1)
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(x2_train, y2_train)
y2_pred = model.predict(x2_test)
```

```python
score = r2_score(y2_test, y2_pred)
round(score,2)
```

# Question 13

```python
#separate target from data for which we have to predict values
x = df.drop(['Appliances'], axis=1)
y = df['Appliances'] 
```

```python
# now split data into two parts one for training model and second for testing model
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model2 = LinearRegression()
model2.fit(x_train, y_train)
y_pred = model2.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
round(mae, 2)
```

# Question 14

```python
rss = np.sum(np.square(y_test - y_pred))
round(rss, 2)
```

# Question 15

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
round(rmse, 3)
```

# Question 16

```python
# Coefficient of Determination (r2 score)
r2score = r2_score(y_test, y_pred)
round(r2score, 2)
```

# Question 17

```python
weights = model2.coef_
features = df.columns[1:]
for i,w in enumerate(weights):
  print('%s,\t\t Weight: %f' % (features[i], w))
```

# Question 18

```python
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)

y_ridge_pred = ridge_reg.predict(x_test)
rmse2 = np.sqrt(mean_squared_error(y_test, y_ridge_pred))
round(rmse2, 3)
```

# Question 19

```python
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

y_lasso_pred = lasso_reg.predict(x_test)

print('No. of non-zero feature weights : ',(lasso_reg.coef_!=0).sum())
```

# Question 20

```python
rmse3 = np.sqrt(mean_squared_error(y_test, y_lasso_pred))
print('RMSE : ',round(rmse3, 3))
```
